```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import xarray as xr

print('xarray version: %s' % xr.__version__)

sns.set_style('whitegrid')
sns.set_context('paper')
```

```python
ds = xr.open_dataset('../data/ERA-I_surface_20160501.nc')
```

```python
ds_mean = ds.mean('time')
```

```python
fig, ax = plt.subplots(figsize=(7.5, 3))
mappable = (ds.isel(time=0).t2m - 273.15).plot.pcolormesh(vmin=-60, vmax=40,ax=ax,
                                                          add_colorbar=False, cmap='viridis')
ax.set_aspect('equal')
cbar = plt.colorbar(mappable, pad=0.02, extend='both')
cbar.set_label('Surface Air Temperature (C)')

fig.savefig('../figures/ERA-Interim_surface_air_temp_figure.png', dpi=300, bbox_inches='tight')
```
